```python

import subprocess
import numpy as np
import pandas as pd
```

```python
dfUrls = pd.read_csv("urls_to_get.csv")
dfUrls.head()
```

```python
for i,row in dfUrls.iterrows():
  print(row['urls'])
  subprocess.run(['wget',row['urls']])
  subprocess.run(['unzip',row['urls'].split("/")[-1]])
```

```python
result = subprocess.run(["ls"],capture_output=True,text=True)
print(result.stdout)
```
